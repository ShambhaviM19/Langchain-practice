```python
from langchain_community.embeddings import OllamaEmbeddings
embeddings=(
    OllamaEmbeddings(model="llama3.1")
)
```

```python
result=embeddings.embed_documents([
    'I have not always loved solving maths',
    'My JEE teacher inspired me and I love maths now'
])
print(result)
```

```python
embeddings.embed_query("My JEE teacher made me love maths")
```
